```python
#initializing x "exploratory variable"
x=iris.iloc[:,0:4]
x
```

```python
#initializing y "categorical variable"
y=iris.iloc[:,4]
y
```

**SPLITTING THE DATA**

```python
#importing train test and split from sklearn
from sklearn.model_selection import train_test_split
#spliting the data
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=1)
#shapes of x_train and y_train
x_train.shape,y_train.shape
```

```python
#shapes of x_test and y_test
x_test.shape,y_test.shape
```

```python
plt.figure(figsize=(8,8))
plt.pie(iris['class'].value_counts().values,labels=iris['class'].value_counts().values)
plt.show()
```

```python
pd.plotting.scatter_matrix(iris,figsize=(12,8))
```

```python
def plot_kde(a):
    facet=sb.FacetGrid(iris,hue='class',aspect=4,palette='PuBu')
    facet.map(sb.kdeplot,a,shade=True)
    facet.set(xlim=(0,iris[a].max()))
    plt.title(a.title())
    plt.show()
```

```python
for col in iris.columns[:4]:
    plot_kde(col)  
```

```python
from sklearn.neighbors import KNeighborsClassifier
model_kn=KNeighborsClassifier()
#training the data using fit method
model_kn.fit(x_train,y_train)
```

```python
#prediction using kneighbors classsifier
y_kn_pred=model_kn.predict(x_test)
y_kn_pred
```
